## Step 2 Data Cleaning

### Check Missing Values

```python
df.isnull().sum()
```

### Initial Insight

From the initial dataset exploration, we observe that the dataset contains several columns related to sales transactions. However, the dataset also contains missing values and inconsistent entries which must be cleaned before performing detailed analysis.

### Display Rows with Missing Values

```python
df[df.isnull().any(axis=1)]
```

### Handle Missing Values (Numerical Columns)

For numerical columns like:

- Sales

- Profit

- Discount

We fill missing values using mean ,median.

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())


```

```python
df["Profit"] = df["Profit"].fillna(df["Profit"].mean())
```

```python
df["Discount"] = df["Discount"].fillna(df["Discount"].mean())
```

### Handle Missing Values (Categorical Columns)

For columns like City.

- Fill with most common value.

```python
df["City"] = df['City'].fillna(df['City'].mode()[0])
```

```python
df.isnull().sum()
```

### Check Duplicate Rows

```python
df.duplicated().sum()
```

### Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

```python
# Verify again:

df.duplicated().sum()
```

### Fix Inconsistent Region Names

- Dataset contains values like:

East
east
WEST
West

```python
df['Region'] = df['Region'].str.title()
```

```python
# Now check unique values.

df['Region'].unique()
```

### Fix Category Column

- Some rows contain:

Furniture

furniture

```python
df['Category'] = df['Category'].str.title()
```

### Convert Order Date to Datetime

Currently Order_Date is string.

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
df.dtypes
```

### Extract Year and Month

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

### Final Check

```python
df.info()
```

Now dataset should be:

Clean

- No missing values

- Correct data types

- Consistent categories

### Insight

After cleaning the dataset:

- Missing values were successfully handled.

- Duplicate records were removed.

- Inconsistent text values were standardized.

- Date columns were converted into proper datetime format.

- The dataset is now ready for data analysis and feature exploration.

## STEP 8 — Apply & Lambda Functions in Pandas

### Using apply() on a Column

Example: Increase Sales by 5%

```python
df['Updated_Sales'] = df['Sales'].apply(lambda x: x * 1.05)
```

Explanation:

- apply() → applies a function to each value

- lambda → small anonymous function

- x → each value in Sales column

### Create Profit Category

Example: Categorize profit as High or Low.

```python
df['Profit_Category'] = df['Profit'].apply(lambda x: 'High Profit' if x > 200 else 'Low Profit')
```

### Apply Function on Multiple Columns

Example: Calculate Net Revenue

```python
df['Net_Revenue'] = df.apply(lambda row: row['Sales'] - row['Discount'], axis=1)
```

Explanation:

- axis=1 → row-wise operation

- Uses multiple columns.

### Use apply() for String Operations

Example: Convert City names to uppercase.

```python
df['City'] = df['City'].apply(lambda x: x.upper())
```

### Create Discount Category

```python
df['Discount_Level'] = df['Discount'].apply(
    lambda x: 'High Discount' if x > 0.2 else 'Low Discount'
)
```

### Apply Mathematical Transformation

Example: Round Sales values.

```python
df['Sales'] = df['Sales'].apply(lambda x: round(x,2))
```

### Use apply() with Conditional Logic

Example: Identify large orders.

```python
df['Large_Order'] = df['Quantity'].apply(lambda x: 'Yes' if x >= 5 else 'No')
```

### Apply Function to Entire DataFrame

Example: Find maximum values per column.

```python
df.apply(np.max)
```

### Replace Values

Example: Replace category name.

```python
df['Category'] = df['Category'].replace('Office Supplies','Office')
```

### Insights

Using apply() and lambda functions allows flexible data transformation. These operations help create new features such as profit categories, discount levels, and large order indicators, which can be useful for deeper business analysis and machine learning models.
